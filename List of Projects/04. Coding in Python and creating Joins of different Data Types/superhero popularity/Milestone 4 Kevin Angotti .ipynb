```python
SuperHero = requests.get(format(url))
SuperHero
```

```python
SuperHero_json = SuperHero.json()
SuperHero_json
```

```python
milestone4 = pd.DataFrame(SuperHero_json)
```

```python
milestone4
```

```python
# Step 1 Unpacking the differnt categories
# powerstats + ading in an extra column to sum the powerstats 
powerstats = milestone4['powerstats'].apply(pd.Series)
powerstats['attribute_score'] = powerstats.apply(lambda x: x['intelligence'] + x['strength']+ x['speed']+ x['durability'] + x['power']+ x['combat'], axis=1)
# creates a new variable to hold the summary of all the power abilities. will be used as a test variable for analysis 
powerstats
```

```python
# appearance
appearance = milestone4['appearance'].apply(pd.Series)
appearance
```

```python
# biography
biography = milestone4['biography'].apply(pd.Series)
biography
```

```python
# work
work = milestone4['work'].apply(pd.Series)
work
```

```python
# connections
connections = milestone4['connections'].apply(pd.Series)
connections
```

```python
# Step 2 Append the data frames 
milestone4 = pd.concat([milestone4, powerstats, appearance, biography, work], axis=1)
```

```python
milestone4
```

```python
# Step 3 drop columns that are not needed
milestone4 = milestone4.drop(columns='powerstats')
milestone4 = milestone4.drop(columns='appearance')
milestone4 = milestone4.drop(columns='biography')
milestone4 = milestone4.drop(columns='work')
milestone4 = milestone4.drop(columns='connections')
milestone4 = milestone4.drop(columns='images')
milestone4.head()
```

```python
# Step 4 checks for duplicates 
dups = milestone4.pivot_table(index=['slug'], aggfunc='size')
dups.head(60)
```
