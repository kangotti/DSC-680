```python
Milestone2_df = pd.read_csv("SuperHeroDataset.csv", sep=",", header=None, engine='python', encoding="utf-8-sig") # imports the dataset
Milestone2_df
```

```python
# Create a new variable called 'header' from the first row of the dataset
# This calls the first row for the header
header = Milestone2_df.iloc[0]
# take the rest of your data minus the header row
Milestone2_df = Milestone2_df[1:]
# set the header row as the df header
Milestone2_df.columns = header
Milestone2_df
```

```python
Milestone2_df.rename(columns = {'Name':'name'}, inplace = True)
```

```python
Milestone2_df
```

```python
print(Milestone2_df.dtypes)
```

```python
# changes all the ability variables to a numeric type from an object type. 
Milestone2_df['intelligence_score'] = pd.to_numeric(Milestone2_df['intelligence_score'], errors='coerce')
Milestone2_df['strength_score'] = pd.to_numeric(Milestone2_df['strength_score'], errors='coerce')
Milestone2_df['speed_score'] = pd.to_numeric(Milestone2_df['speed_score'], errors='coerce')
Milestone2_df['durability_score'] = pd.to_numeric(Milestone2_df['durability_score'], errors='coerce')
Milestone2_df['power_score'] = pd.to_numeric(Milestone2_df['power_score'], errors='coerce')
Milestone2_df['combat_score'] = pd.to_numeric(Milestone2_df['combat_score'], errors='coerce')
```

```python
print(Milestone2_df.dtypes[6:13])
```

```python
Milestone2_df['attribute_score'] = Milestone2_df.apply(lambda x: x['intelligence_score'] + x['strength_score']+ x['speed_score']+ x['durability_score'] + x['power_score']+ x['combat_score'], axis=1)
# creates a new variable to hold the summary of all the power abilities. will be used as a test variable for analysis 
```
