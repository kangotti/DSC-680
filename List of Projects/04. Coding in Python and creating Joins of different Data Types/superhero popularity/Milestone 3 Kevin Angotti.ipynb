```python
import pandas as pd
import numpy as np
import math
import requests
import matplotlib as plt
import csv
from scipy import stats
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/List_of_highest-grossing_superhero_films'   # url for html data
resp=requests.get(url)
if resp.status_code==200:
    print("Successfully opened page")  # prnts success is able to connect to site
soup=BeautifulSoup(resp.text, 'html.parser') # prints the raw html data and parses 
    
   
print(soup.prettify()) # prints the data
```

```python
total_tables = soup.find_all("table") # locates tables on the site
print("The total number of tables = {} ".format(len(total_tables))) # prints how many tables are available 
```
